```python
import pandas as pd
```

DataFrame을 만들기 위해 파이썬 dictionary를 다음과 같이 정의한다.

```python
people = {
    "이름" : ["길동", "춘향", "몽룡"],
    "성" : ["홍", "성", "이"],
    "이메일" : ["gdhong@gmail.com", "chsung@gmail.com", "mrlee@gmail.com"]
}
```

people로 부터 DataFrame을 생성하고 그 놈을 df라 하자

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
type(df)
```

last칼럼값을 오름차순으로 정렬하자. sort_values()함수 이용.

```python
df.sort_values(by='성')
```

```python
df
```

기본적으로 sort_values는 오름차순 정렬한다. sort_values(ascending=True)가 기본 값
내림차순으로 정렬하고자 한다면

```python
df.sort_values(by='성', ascending=False)
```

SQL에서도 order by 절에 정렬 기준을 여러개 명시할 수 있다.
pandas도 그렇다

```python
df.sort_values(by=['성','이름'], ascending=[True,False])
```

pandas에서 상위 n개의 결과를 볼 수 있는 nlargest() 함수를 제공한다. 상위 10명의 연봉만 보고 싶다

```python
df['SalaryUSD'].nlargest(10)
```

연봉 top 10의 다른 정보도 알고 싶다. 그러면?

```python
df.nlargest(10, 'SalaryUSD')[['Country','EdLevel','SalaryUSD']]
```
