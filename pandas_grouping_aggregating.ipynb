aggregation : 데이터의 여러 조각들을 조합해 하나의 결과를 생성하는 것
예를 들어, 조사에 응한 개발자들의 평균연봉에 대한 정보를 알고 싶다면,
관련 aggregate 함수를 사용할 수 있다.

central tendency : average (= mean), median

DB를 예를 들면 : select count(*) from members;
              select email, name from members;
              select email, name from members where age > 25
              select dept, count(*) from members group by dept; (O)
              select dept, age, count(*) from members group by dept; (X)
              select dept, count(age) from members group by dept; (O)
DB에서 aggregate function (집단함수) : count, min, max, average, sum

각 나라의 물가 등에 따라 연봉의 가치가 다르니, 각 나라별 연봉의 평균을 구해볼 수 있다.
각 나라별 어떤 소셜 미디어가 인기가 있는지 알고 싶다면...
groupby 함수 사용법을 알아야 한다.
pandas 문서를 보면 groupby 연산은 3단계로 구성된다.
    1. Split, 2. Apply Function, 3. Combine results

```python
country_grp = df.groupby(['Country'])
```

```python
country_grp.get_group('United States')
```

```python
country_grp.get_group('India')
```

```python
country_grp['SocialMedia'].value_counts().head(50)
```

```python
country_grp['SocialMedia'].value_counts()
```

```python
country_grp['SocialMedia'].value_counts().loc['South Korea']
```

```python
country_grp['SocialMedia'].value_counts().loc[['South Korea','Republic of Korea']]
```

```python
filt = df['Country'] == 'India'
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'SocialMedia'].value_counts()
```

러시아(Russian Federation) 에서 어던 소셜미디어가 인기가 있는지 정규화(normalize)해서 표시하시요.

```python
filt = df['Country'] == 'Russian Federation'
df.loc[filt, 'SocialMedia'].value_counts(normalize=True)
```

```python
country_grp['SocialMedia'].value_counts(normalize=True).loc['Russian Federation']
```

응답자들의 연봉의 median 값을 저 위에서 알아 봤고..
이제는 각 나라별 연봉의 median 값을 알아보자.

```python
country_grp['ConvertedComp'].median().head(50)
```

```python
country_grp['ConvertedComp'].median().sort_values(ascending=False)
```

```python
country_grp['ConvertedComp'].median().loc['South Korea']
```

```python
country_grp['ConvertedComp'].mean()
```

각 나라별 연봉의 중간값과 평균을 한 번에 보고싶다.

그러면, aggregate 함수를 여러개 적용해야 하는데... 이 때

agg() 함수를 사용하면 된다.

```python
country_grp['ConvertedComp'].agg(['median', 'mean'])
```
